```python
# feature engineering - 'region variable'
```

```python
df_food['city'].isnull().sum()
```

```python
df_food['city'].value_counts()
```

```python
def assign_region(city):
    if city in ['Delhi','Noida','Greater Noida','Gurgaon']:
        return 'North'
    elif city in ['Mumbai','Pune']:
        return 'West'
    elif city in ['Chennai','Bangalore','Hyderabad']:
        return 'South'
    else:
        return 'Other'

if 'city' in df_food.columns:
    df_food['region'] = df_food['city'].apply(assign_region)
    print(df_food[['city','region']].tail())
```
